You might use the phrase "It lasted just an instant" to refer to something that lasted for a very shan time interval. But in physics an instant has no duration at all; it refers to a single value of time

```python
from scipy.special import fresnel
  
tO=-3
tF=3

t = np.linspace(tO, tF, 100)
y, x = fresnel(t)


plt.plot(x, y,'ro--',ms=4,lw=2)
plt.title('Trajectory') # title
plt.xlabel('x(in m)')              # legend of the abscissa  "ex" axis
plt.ylabel('y(in m)')              # legend of the ordinate "ey" axis
plt.axis('equal')                  # same sclale for both axis
```

```python
### BEGIN SOLUTION
from scipy.special import fresnel
  
tO=-3
tF=3
npoints=100
D=0

t = np.linspace(tO, tF, npoints) # The time is s
y, x = fresnel(t) # The lenght is m

for ti in np.arange(0,npoints-1,1):
    dx2 = (x[ti+1]-x[ti])**2
    dy2 = (y[ti+1]-y[ti])**2
    dl  = (dx2+dy2)**0.5
    D += dl

print("The lenght of the path is L = {:0.3f} m".format(D))

### END SOLUTION
```
